```python
import pandas as pd
```

```python
#load df
df = pd.read_csv('../data/ufc-master.csv')
```

```python
display(df)
```

```python
def get_fight_time_secs(r, t):
    if r== '' or t == '':
        return ''
    else:
        calculated_time = 300 * (r-1)
        t_split = str(t).split(':')
        #Check for nan
        if t_split[0] != 'nan':
            #print(t_split[0])
            calculated_time = calculated_time + 60 * int(t_split[0]) + int(t_split[1])
    return calculated_time
```

```python
df['fight_time_secs'] = df.apply(lambda x: get_fight_time_secs(x['round'], x['time']), axis=1)
```
